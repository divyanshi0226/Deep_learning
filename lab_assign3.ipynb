```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv(r"C:\Users\divya\Downloads\archive (5)\multiple_linear_regression_dataset.csv")
df.head()
```

```python
X=df[['age','experience']]
Y=df['income'].values
```

```python
X_mean=X.mean(axis=0)
X_std=X.std(axis=0)
X=(X-X_mean)/X_std
```

```python
m=X.shape[0]
X=np.c_[np.ones(m),X]
```

```python
theta=np.zeros(X.shape[1])
alpha=0.01
iterations=1000
```

```python
for _ in range(iterations):
    predictions=X.dot(theta)
    errors=predictions-Y

    gradients=(1/m)*X.T.dot(errors)
    theta=theta-alpha*gradients
```

```python
print("intercept",theta[0])
print("coefficients",theta[1:])
```

```python
y_pred=X.dot(theta)
```

```python
mse=np.mean((Y-y_pred)**2)
rmse=np.sqrt(mse)

ss_total=np.sum((Y-np.mean(Y))**2)
ss_res=np.sum((Y-y_pred)**2)
r2=1-(ss_res/ss_total)
```

```python
print("rmse",rmse)
print("r2",r2)
```

```python
plt.scatter(Y,y_pred)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.show()
```

```python
m, n = X.shape
X = np.c_[np.ones(m), X]   # add bias column
```

```python
weights = np.zeros(n + 1)   # including bias
learning_rate = 0.01
epochs = 1000
```

```python
for epoch in range(epochs):
    for i in range(m):
        y_pred = np.dot(X[i], weights)   # linear activation
        error = Y[i] - y_pred
        
        # weight update rule
        weights = weights + learning_rate * error * X[i]
```

```python
print("Intercept (Bias):", weights[0])
print("Coefficients:", weights[1:])
```

```python
mse = np.mean((Y - y_pred) ** 2)
rmse = np.sqrt(mse)

ss_total = np.sum((Y - np.mean(Y)) ** 2)
ss_res = np.sum((Y - y_pred) ** 2)
r2 = 1 - (ss_res / ss_total)

print("RMSE:", rmse)
print("R² Score:", r2)
```
